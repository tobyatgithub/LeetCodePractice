```python
# version 1, idea: we cut the string into length = 2n-3 cells, and do the same thing to each cell
class Solution:
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        if numRows == 1:
            return s
        
        
        output = {}
        # output.setdefault(1, [])
        # also, i think it's fair to assume numRows > 1
        interval = 2 * numRows - 2
        i = 0
        # for n in range(numRows):
        #     output.setdefault(n, [])
        while (i+interval) <= len(s):
            tmp = list(s[i:i+interval])
            # print(tmp)
            for n in range(1,numRows+1):
                output.setdefault(n, [])
                if (n == 1) or (n == numRows): 
                    output[n].append(tmp.pop(0))
                    # print("tmp = ", tmp)
                else:
                    # print("n = ", n)
                    output[n].append(tmp.pop(0))
                    output[n].append(tmp.pop(-1))
        #     break
            i = i + interval

        # last case:
        if (i < len(s)):
            # print("last case")
            tmp = list(s[i:])
            tmp = tmp + [""] * (interval - len(s) + i)
            # print(tmp)
            for n in range(1,numRows+1):
                output.setdefault(n, [])
                if (n == 1) or (n == numRows): 
                    output[n].append(tmp.pop(0))
                    # print("tmp = ", tmp)
                else:
                    output[n].append(tmp.pop(0))
                    output[n].append(tmp.pop(-1))    
        r = []
        for i in output.values():
            r += i
        
        print("".join([_ for _ in r if _ != ""]))
        return "".join([_ for _ in r if _ != ""])
```

```python
# version 2, from discussion, smart way to simulate zig zag moving process!
# the core idea is:
# index starting from 0, all the way to 1, 2, ... , end (which = numRows-1, since python start from 0)
# when index hit bottom, now we read the string backwards (thus index += -1)
# when we hit all the way back to 0 (first line,) we start to read forward again (index += 1)

# class Solution(object):
def convert(s, numRows):
    """
    :type s: str
    :type numRows: int
    :rtype: str
    """
    if numRows == 1 or numRows >= len(s):
        return s

    L = [''] * numRows
    index, step = 0, 1

    for x in s:
        print(s)
        print("running x = ", x)
        print("running, L = ", L)
        print("index = %d, step = %d" %(index,step))
        L[index] += x
        if index == 0:
            step = 1
        elif index == numRows -1:
            step = -1
        index += step
        print("after this round, L = ", L)
        print("index = %d, step = %d" %(index,step))
        print("\n")

    return ''.join(L)
```
